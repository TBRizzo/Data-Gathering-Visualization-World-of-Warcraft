# IMPORTS TOP 0.1% PLAYERS FROM EACH SEASON

### Link

https://raider.io/mythic-plus-character-rankings/season-7.2.0/world/all/all/0#content

```python
#libs imports
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import math
import re
```

## Data colecting

```python
#empty lists to store desired data
role = []
season = []
wowclass = []
region = []
#list with all the seasons
seasons = ['season-7.2.0', 'season-7.3.0', 'season-bfa-1', 'season-bfa-2', 'season-bfa-3', 'season-bfa-4', 'season-sl-1', 'season-sl-2', 'season-sl-3', 'season-sl-4']
#list with roles (to extract players from the proper 'role' url)
avail_roles = ['tank', 'healer', 'dps']

#loop for each role
for r in avail_roles:
    #loop for the seasons
    for s in seasons:
        url = f'https://raider.io/mythic-plus-character-rankings/{s}/world/all/{r}/0#content'
        html = requests.get(url).content
        soup = BeautifulSoup(html)
        last_link = soup.find_all('a', {'class':'rio-pagination--button'})[1].attrs['href'] #finding the max number of pages
        pattern = '\d{3,}'
        #setting the extraction loop range, based on 0.1% of the max number of pages (which approximately also means 0.1% of players)
        loop_range = math.floor(int(re.findall(pattern, last_link)[0])*0.001) + 1 
        #loop for player data extraction
        for spage in range(loop_range):
            url_curr = f'https://raider.io/mythic-plus-character-rankings/{s}/world/all/{r}/{spage}#content'
            html_curr = requests.get(url_curr).content
            soup_curr = BeautifulSoup(html_curr)
            #rng is the number of values on the rank 'table' in each page (in case there was less than 20 players in a page)
            rng = len(soup_curr.find_all('table', attrs={'class':'slds-table slds-table--bordered slds-table--striped rio-rankings-table'})[0].find_all('tr'))
            for i in range(1, rng):
                #print for code progress verification and error finding
                print(f'season:{s} page:{spage} item:{i}')
                role.append(r)
                season.append(s)
                wowclass.append(soup_curr.find_all('div', attrs={'class':'slds-tile slds-media'})[i-1].find('div', attrs={'class':'sprite-scaler'}).contents[0].attrs['class'][0])
                #print for code progress verification and error finding
                print(soup_curr.find_all('div', attrs={'class':'slds-tile slds-media'})[i-1].find('div', attrs={'class':'sprite-scaler'}).contents[0].attrs['class'][0])
                region.append(re.findall('[A-Z][A-Z]', soup_curr.find_all('div', attrs={'class':'slds-tile__detail slds-text-body--small'})[i-1].text)[0])
```

## Creating the dataframe

```python
col_names = ['Role', 'Season', 'Class', 'Region']


df = pd.DataFrame({'Role': role,'Season': season,'Class': wowclass,'Region': region})
df.head()
```

```python
#function to change class names for better visualization
def classwow(x):
    if "class_warrior" in x:
        return "Warrior"
    elif "class_paladin" in x:
        return "Paladin"
    elif "class_hunter" in x:
        return "Hunter"
    elif "class_rogue" in x:
        return "Rogue"
    elif "class_priest" in x:
        return "Priest"
    elif "class_death-knight" in x:
        return "Death Knight"
    elif "class_shaman" in x:
        return "Shaman"
    elif "class_mage" in x:
        return "Mage"
    elif "class_warlock" in x:
        return "Warlock"
    elif "class_monk" in x:
        return "Monk"
    elif "class_druid" in x:
        return "Druid"
    elif "class_demon-hunter" in x:
        return "Demon Hunter"
    else:
        return "Evoker"
```

```python
#function to change season names for better visualization
def seasonwow(x):
    if "season-7.2.0" in x:
        return "Legion 7.2.0"
    elif "season-7.3.0" in x:
        return "Legion 7.3.0"
    elif "season-bfa-1" in x:
        return "BfA Season 1"
    elif "season-bfa-2" in x:
        return "BfA Season 2"
    elif "season-bfa-3" in x:
        return "BfA Season 3"
    elif "season-bfa-4" in x:
        return "BfA Season 4"
    elif "season-sl-1" in x:
        return "SL Season 1"
    elif "season-sl-2" in x:
        return "SL Season 2"
    elif "season-sl-3" in x:
        return "SL Season 3"
    else:
        return "SL Season 4"
```

```python
#created a new variable to avoid spoiling the original dataframe
df_2 = df
df_2.head()
```

```python
df_2['Season'] = df_2['Season'].map(seasonwow)

df_2.head()
```

```python
df_2['Class'] = df_2['Class'].map(classwow)

df_2.head()
```
